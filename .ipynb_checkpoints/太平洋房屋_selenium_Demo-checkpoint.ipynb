```python
from selenium.webdriver.common.by import By
import selenium.webdriver as webdriver
import time
from bs4 import BeautifulSoup

driver = webdriver.Firefox()
driver.get("http://www.pacific.com.tw")

ele = driver.find_element_by_id("ctl00_ctl00_Main_Main_ctl00_HouseSearch1_TabContainer1_TabPanel11_ddlAddr1")

print type(ele.find_elements_by_tag_name('option')) #list
print ele.find_elements_by_tag_name('option')[1].text #台北市
ele.find_elements_by_tag_name('option')[1].click() #選台北市
   
driver.find_element_by_id("ctl00_ctl00_Main_Main_ctl00_HouseSearch1_TabContainer1_TabPanel11_LinkButton1").click() #按-開始搜尋

time.sleep(2)
print driver.current_url
getHTML(BeautifulSoup(driver.page_source))
driver.find_element_by_id("ctl00_ctl00_Main_Main_ImageButtonNext").click()

time.sleep(2)
print driver.current_url
getHTML(BeautifulSoup(driver.page_source))
driver.find_element_by_id("ctl00_ctl00_Main_Main_ImageButtonNext").click()

driver.close()
```
